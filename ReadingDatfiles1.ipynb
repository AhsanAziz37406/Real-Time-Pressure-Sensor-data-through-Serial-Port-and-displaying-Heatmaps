```python
# Count finding the Header "fafa or fa fa" form each files
def count_header(data):
  tail = 0
  for i in range(0, len(data), 4):
    byte = data[i:i+4]
    # print(byte)
    if byte == 'fafa':
      # print(byte, i)
      tail += 1
  # print(data[44:44+2])

  return tail
```

```python
# Count finding the Tail "fefe or fe fe" form each files
def count_tail(data):
  tail = 0
  for i in range(0, len(data), 4):
    byte = data[i:i+4]
    # print(byte)
    if byte == 'fefe':
      # print(byte, i)
      tail += 1
  # print(data[44:44+2])

  return tail
```

```python
#function for count all item from dat files
def count_values(data):
  return (len(data)/2) - count_header(data)
  
```
